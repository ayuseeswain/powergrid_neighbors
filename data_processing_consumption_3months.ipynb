This file will create an csv-file from the consumption.csv that has the same time_interval as the production.csv, since this only covers 3 months and consumption.csv covers 1 year.

```python
import pandas as pd
```

```python
source_file_cons = 'data/consumption.csv'
source_file_prod = 'data/pv_production_3months.csv'
target_file_cons_red = 'data/consumption_3months.csv'

# Shared year 
shared_year = 2020
```

```python
# Consumption data
df_cons = pd.read_csv(source_file_cons)
df_cons['datetime'] = pd.to_datetime(df_cons['DateTime'])
df_cons.drop(columns=['Unnamed: 0','DateTime'],inplace=True)

# Change the year to shared year
timestamps_list = []
for i in range(df_cons.shape[0]):
    t = df_cons['datetime'].iloc[i]
    timestamps_list.append(pd.Timestamp(shared_year,t.month,t.day,t.hour,t.minute))
timestamps = pd.DataFrame(timestamps_list,columns=['times'])
df_cons['datetime'] = timestamps['times']

# Production data
df_prod = pd.read_csv(source_file_prod)
df_prod['datetime'] = pd.to_datetime(df_prod['datetime'])

# Change year to shared year
timestamps_list = []
for i in range(df_prod.shape[0]):
    t = df_prod['datetime'].iloc[i]
    timestamps_list.append(pd.Timestamp(shared_year,t.month,t.day,t.hour,t.minute))
timestamps = pd.DataFrame(timestamps_list,columns=['times'])
df_prod['datetime'] = timestamps['times']
```

```python
# Take only rows in consumption dataframe that have the same datetime
df_cons_red = df_cons.where(df_cons['datetime'].isin(df_prod['datetime'])).dropna()

# Set as index
df_cons_red = df_cons_red.set_index(df_cons_red['datetime'])
# Store the reduced consumption DataFrame to a new file
df_cons_red.to_csv(target_file_cons_red)
```
